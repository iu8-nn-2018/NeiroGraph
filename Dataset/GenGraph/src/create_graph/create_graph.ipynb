## Creating a graph

```python
import networkx as nx
G = nx.Graph()
```

## Nodes

```python
G.add_node(1)
nx.draw(G)
```

```python
G.add_nodes_from([2, 3])
nx.draw(G)
```

```python
H = nx.path_graph(10)
G.add_nodes_from(H)
nx.draw(G)
```

## Edges

```python
G.add_edge(1, 2)
nx.draw(G)
```

```python
e = (2, 3)
G.add_edge(*e)
nx.draw(G)
```

```python
G.add_edges_from([(1, 2), (1, 3)])
nx.draw(G)
```

```python
G.add_edges_from(H.edges)
nx.draw(G)
```

```python
G.clear()

G.add_edges_from([(1, 2), (1, 3)])
G.add_node(1)
G.add_edge(1, 2)
G.add_node("spam")
G.add_nodes_from("spam")
G.add_edge(3, 'm')
nx.draw(G)
```
